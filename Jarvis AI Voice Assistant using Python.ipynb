```python
# @Amit Sharma
# Importing required library

import pyttsx3
import speech_recognition as sr 
import webbrowser
import datetime
import pyjokes
import os
import time

# Speech to text

def sptext():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        recognizer.adjust_for_ambient_noise(source)  # noise reduction
        audio = recognizer.listen(source)
        try:
            print("Recognizing...")
            data = recognizer.recognize_google(audio)
            print(data)
            return data
        except sr.UnknownValueError:
            print("Not understood.")
            return ""

# Text to speech

def speechtx(x):
    engine = pyttsx3.init()  # creating object
    voices = engine.getProperty('voices')
    engine.setProperty('voice', voices[0].id)  # voices ->0 male, 1 female
    engine.setProperty('rate', 150)  # voice speed
    engine.say(x)
    engine.runAndWait()

speechtx("Hello, Welcome to Jarvis A.I. How can i help you")  # Jarvis will speak this text

if __name__ == '__main__':

    if sptext().lower() == "Jarvis":  # <- say jarvis to start recognizing
        while True:
            data1 = sptext().lower()

            if "your name" in data1: 
                name = "My name is jarvis. I am created by Amit Sharma."
                speechtx(name)

            elif "old are you" in data1:
                age = "I am two years old."
                speechtx(age)

            elif 'time' in data1:
                current_time = datetime.datetime.now().strftime("%I:%M %p")  # hrs, min, am/pm
                speechtx(f"The time is {current_time}")

            elif 'youtube' in data1:  # <- open YouTube
                webbrowser.open("https://www.youtube.com/")

            elif 'web' in data1:  # <- open Google
                webbrowser.open("https://www.google.com/")

            elif 'github' in data1:  # <- open GitHub
                webbrowser.open("https://github.com/Amit123456777")

            elif "joke" in data1: # <- play joke
                joke_1 = pyjokes.get_joke(language="en", category="neutral")
                speechtx(joke_1)
                print(joke_1)

            elif 'play song' in data1:  # <- play a song
                add = "C:\\Users\\Acer\\Downloads\\Songs"  # <- song folder
                listsong = os.listdir(add)
                print(listsong)
                os.startfile(os.path.join(add, listsong[0]))  # Play the first song

            elif "exit" in data1:  # <- exit the program
                speechtx("Thank you!")
                break

            time.sleep(5)  # Pause for 5 seconds before listening again

    else:
        print("Thanks!")  # If "hey jarvis" is not said, print "Thanks"
```

```python
####### How to use this code
# #<-------------------------------->

# run code -> when listening... -> say "Jarvis" -> wait for recognition -> now you can give the command

# Step 1:
# Listening... <- First, say "Jarvis" to start talking
# Recognizing...
# Jarvis

# Step 2:
# Now try commands like "play song," "what is your name," "open YouTube," "open GitHub" (more are mentioned in the code).
# Listening... <- e.g., "Jarvis, open YouTube"
# Recognizing... <- wait for recognition
# open YouTube <- output

# Step 3:
# Listening... <- Continues listening due to the infinite loop
# Recognizing...
# play song

# Step 4:
# Listening... <- Say "exit" to stop
# Recognizing...
# exit

```
